```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('localhost', 27018)
db = client.TableView
collection = db["dbSNP153_Annotation"]
```

```python
db.list_collection_names()
```

```python
df = pd.read_csv("dbSNP153_chr6_NHCFV_AF_HiC_AnnotationResults.txt", sep="\t",  encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
col= df.columns
length = len(col)
```

```python
for i,r in df.iterrows():
    collection.update_one({"RSID": r.RSID, "Alt":r.Alt, "End":r.End}, {"$set":{col[i]:r[col[i]] for i in range(6,length)}})
    if i %100000==0:
        print(i)
```
