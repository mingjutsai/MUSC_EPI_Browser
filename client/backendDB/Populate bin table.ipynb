```python
import pandas as pd
from pymongo import MongoClient
```

```python
# connect to workstation mongoDB, set db to the table view, the collection to bin
client = MongoClient('localhost', 27017)
db = client.TableView
collection = db["bin_NHCFV"]
```

```python
df_bin = pd.read_csv("binTable.NHCFV", sep="\t")
```

```python
df_bin.head()
```

```python
df_bin[df_bin["HiC Interactions"] != "."]
```

```python
for i, r in df_bin.iterrows():
    document = {}
    document["Chr"] = r["#Chromosome"]
    document["Start"] = r["Start"]
    document["End"] = r["End"]
    document["HiC_interactions"] = r["HiC Interactions"].split(";")
    document["ATAC"] = r["ATAC"].split(";")
    document["ChromHMM"] = r["ChromHMM"].split(";")
    document["Footprinting"] = r["Footprinting"].split(";")
    collection.insert_one(document)
```

```python
collection.create_index("Start")
```
